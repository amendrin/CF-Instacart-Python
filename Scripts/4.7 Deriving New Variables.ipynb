# 03. User-Defined Functions

```python
#create subset of first million rows of datafram
df= ords_prods_merged[:1000000]
```

```python
#define function
def price_label(row):
    if row['prices']<=5:
        return 'Low-range product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices']>15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#apply new function
df['price_range']= df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna= False)
```

```python
#most expensive product
df['prices'].max()
```

# 04. loc() Function

```python
df.loc[df['prices']>15, 'price_range_loc']= 'High-range product'
```

```python
df.loc[(df['prices']<=15) & (df['prices']>5), 'price_range_loc']= 'Mid-range product'
```

```python
df.loc[df['prices']<=5, 'price_range_loc']= 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna= False)
```

# 06. For-loops

```python
#which days the most orders take place
ords_prods_merged['orders_day_of_the_week'].value_counts(dropna=False)
```

#### 0(Saturday) is the busiest day of the week, and 4(Wednesday) is the slowest day of the week

```python
result= []

for value in ords_prods_merged['orders_day_of_the_week']:
    if value==0:
        result.append('Busiest day')
    elif value== 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
result
```

```python
ords_prods_merged['busiest_day']= result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna=False)
```

## Q2

#### Busiest days of the week: Saturday and Sunday (0/1)
#### Slowest days of the week: Tuesday and Wednesday (3/4)

```python
result2= []

for value in ords_prods_merged['orders_day_of_the_week']:
    if value in [0,1]:
        result2.append('Busiest Days')
    elif value in [3,4]:
        result2.append('Slowest Days')
    else:
        result2.append('Regular Days')
```

```python
result2
```

```python
#create new column: busiest_days
ords_prods_merged['busiest_days']= result2
```

```python
ords_prods_merged['busiest_days'].value_counts(dropna=False)
```

## Q3

#### Find the busiest hours of the day, labeled by 'most, average and fewest orders'. New column name=busiest_period_of_day

```python
ords_prods_merged.head()
```

```python
ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
ords_prods_merged['order_hour_of_day'].shape
```

```python
result3= []

for value in ords_prods_merged['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13]:
        result3.append('Most orders')
    elif value in [3, 4, 2, 5, 1]:
        result3.append('Fewest orders')
    else:
        result3.append('Average orders')
```

```python
result3
```

```python
#new column: busiest_period_of_day indicating the top 5 busiest and slowest hours of the day
ords_prods_merged['busiest_period_of_day']= result3
```

```python
ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False)
```
